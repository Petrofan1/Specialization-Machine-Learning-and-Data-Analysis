```python
import numpy as np 
import scipy
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```python
x_1 = [1, 15]
n_1 = len(x_1)
matrix_1 = np.zeros((n_1, n_1))
vector_1 = np.zeros((n_1, 1))
```

```python
for i in range(n_1):
    for j in range(n_1):
        matrix_1[i][j] = x_1[i] ** j
    vector_1[i] = f(x_1[i])
print(matrix_1)
print(vector_1)
```

```python
sol_1 = scipy.linalg.solve(matrix_1, vector_1)
```

```python
x = np.arange(0, 15, 0.1)
y_func = f(x)
y_pol = sol_1[0] + sol_1[1]*x
```

```python
plt.plot(x, y_func,  x, y_pol)
plt.show()
```
